This project analyzes a synthetic healthcare dataset to identify trends and provide recommendations for cost-effective and patient-satisfactory healthcare plans.


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Set style
sns.set(style="whitegrid")

# Generate synthetic data
np.random.seed(42)
n = 1000
ages = np.random.randint(18, 90, n)
genders = np.random.choice(['Male', 'Female'], n)
conditions = np.random.choice(['Diabetes', 'Hypertension', 'Heart Disease', 'Asthma', 'None'], n)
income_levels = np.random.choice(['Low', 'Medium', 'High'], n, p=[0.4, 0.4, 0.2])
plan_type = np.random.choice(['Basic', 'Standard', 'Premium'], n)
monthly_cost = np.random.normal(loc=300, scale=50, size=n).round(2)
claims_last_year = np.random.poisson(lam=3, size=n)
satisfaction_score = np.random.randint(1, 6, n)

# Create DataFrame
df = pd.DataFrame({
    'Age': ages,
    'Gender': genders,
    'Condition': conditions,
    'Income_Level': income_levels,
    'Plan_Type': plan_type,
    'Monthly_Cost': monthly_cost,
    'Claims_Last_Year': claims_last_year,
    'Satisfaction_Score': satisfaction_score
})

df.head()
```

```python
df.describe(include='all')
```

```python
plt.figure(figsize=(8, 5))
sns.boxplot(x='Plan_Type', y='Monthly_Cost', data=df)
plt.title('Monthly Cost Distribution by Plan Type')
plt.show()
```

```python
avg_satisfaction_by_condition = df.groupby('Condition')['Satisfaction_Score'].mean().sort_values()

avg_satisfaction_by_condition.plot(kind='barh', color='teal', figsize=(8, 5))
plt.xlabel('Average Satisfaction Score')
plt.title('Satisfaction Score by Health Condition')
plt.show()
```

## Recommendations üìå

- **Plan Optimization:** Lower Premium plan costs for Low-income groups where satisfaction is lower.
- **Targeted Improvements:** Focus on services for Asthma and Diabetes patients who show lower satisfaction.
- **Data-Driven Design:** Use correlations between income levels and plan preferences to offer personalized packages.
- **Satisfaction First:** Boost engagement and education for patients with high claim rates and lower satisfaction scores.

These steps can lead to more inclusive, affordable, and high-quality healthcare solutions.
